```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the decline models
qi = 1000  # Initial production rate
Di_exp = 0.2  # Decline rate for exponential
Di_hyperbolic = 0.3  # Decline rate for hyperbolic
b_hyperbolic = 1.0  # Decline exponent for hyperbolic
Di_harmonic = 0.1  # Decline rate for harmonic

# Time array (in years)
time = np.linspace(0, 10, 100)

# Exponential decline model
def exponential_decline(qi, Di, t):
    return qi * np.exp(-Di * t)

# Hyperbolic decline model
def hyperbolic_decline(qi, Di, b, t):
    return qi / (1 + b * Di * t) ** (1 / b)

# Harmonic decline model
def harmonic_decline(qi, Di, t):
    return qi / (1 + Di * t)

# Calculate the production rates for each model
production_exp = exponential_decline(qi, Di_exp, time)
production_hyperbolic = hyperbolic_decline(qi, Di_hyperbolic, b_hyperbolic, time)
production_harmonic = harmonic_decline(qi, Di_harmonic, time)

# Plot the decline curves
plt.figure(figsize=(10, 6))

plt.plot(time, production_exp, label="Exponential Decline", color='blue', linestyle='--', linewidth=2)
plt.plot(time, production_hyperbolic, label="Hyperbolic Decline", color='green', linestyle='-.', linewidth=2)
plt.plot(time, production_harmonic, label="Harmonic Decline", color='red', linestyle='-', linewidth=2)

# Labels and title
plt.title("Decline Curve Analysis: Exponential, Hyperbolic, Harmonic", fontsize=16)
plt.xlabel("Time (years)", fontsize=14)
plt.ylabel("Production Rate (BPD)", fontsize=14)
plt.legend(loc="upper right")
plt.grid(True)

# Display the plot
plt.show()
```
